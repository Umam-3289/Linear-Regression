```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
```

```python
from sklearn.datasets import load_iris
data=load_iris()
data
```

```python
print(data.DESCR)
print(data.data)
print(data.feature_names)
```

```python
df=pd.DataFrame(data.data,columns=data.feature_names)
df["Target"]=data.target
df
```

```python
df["Target"].unique()
# Here we need to drop 2 because we are understanding the concept of Binary classification
df_final=df[df["Target"]!=2]
df_final
df_final.Target.unique()
```

```python
x=df_final.iloc[:,:-1]
y=df_final.iloc[:,-1]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.20,random_state=1)
print(x_train.shape,x_test.shape)
```

```python
from sklearn.linear_model import LogisticRegression
classifier=LogisticRegression()
classifier.fit(x_train,y_train)
```

```python
# Corresponding to the higher probability class is predicted,cutoff is by default is 0.5
y_pred=classifier.predict(x_test)
print(y_pred)
print(classifier.predict_proba(x_test))
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report
print(f"Confusion Matrix: \n {confusion_matrix(y_test,y_pred)}")
print(f"Accuracy: {accuracy_score(y_test,y_pred)}")
print(f"Classification_report: \n {classification_report(y_test,y_pred)}")
```

                                          Above Summary
First two rows of class 0 and 1 >> precison,recall,f1 score and no. of data points of respective class
Macro avg >> for each class take avg >> avg precision of each class 0 and 1 will be macro avg
Weighted avg is used for class imbalance it gives importance to performance of dominant class

```python
# To calculate ROC AUC curve we need to calculate TPR and FPR
y_pred_proba=classifier.predict_proba(x_test)[:,1] # probability for class 1 only that's why we use [:,1]
y_pred_proba
```

```python
# Reciever Operating Curve
# input : y_test,y_pred_proba and gives output fpr,tpr and threshold with different probabilities
from sklearn.metrics import roc_curve,auc
fpr,tpr,threshold=roc_curve(y_test,y_pred_proba)
print(fpr)
print(tpr)
print(threshold)
```

```python
# calculate auc
roc_auc=auc(fpr,tpr)
print(roc_auc)
```

```python
# plot ROC curve
plt.figure(figsize=(8,6))
plt.plot(fpr,tpr,color='darkorange',linewidth=2,label='ROC curve = (%0.2f)'%roc_auc)
plt.plot([0,1],[0,1],linewidth=2,color='navy',linestyle='--')
plt.xlim([0.0,1.0])
plt.ylim([0.0,1.5])
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("Reciever Operating Curve")
plt.legend(loc="lower right")
plt.show()
```

```python
from sklearn.model_selection import KFold
cv=KFold(n_splits=5)
from sklearn.model_selection import cross_val_score
score=cross_val_score(classifier,x_train,y_train,cv=cv,scoring="accuracy")
print(np.mean(score))
```
